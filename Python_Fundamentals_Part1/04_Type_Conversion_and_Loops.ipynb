## 1. What is Type Conversion?

Type conversion means changing one data type into another.

Examples:
- converting a string to an integer
- converting a number to a string
- converting a list to a tuple


```python
# Implicit conversion
x = 10
print(type(x))
y = 2.5
z = x + y
print(z)
print(type(z))

# Explicit conversion
num_str = "25"
num_int = int(num_str)
print(num_int)
print(type(num_int))
```

## 2. Common Conversion Functions

Some common built-in conversion functions are:
- int()
- float()
- str()
- bool()
- list()
- tuple()
- set()


```python
value = "100"
print(int(value))
print(float(value))
print(str(42))
print(bool(0))
print(list("abc"))
print(tuple([1, 2, 3]))
print(set([1, 2, 2, 3]))
```

## 3. for Loops

A for loop is used to iterate over a sequence such as a list, string, or range.

### Example


```python
for i in range(1, 6):
    print(i)

fruits = ["apple", "banana", "mango"]
for fruit in fruits:
    print(fruit)
```

## 4. while Loops

A while loop repeats as long as a condition is True.

### Example


```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

## 5. break, continue, and pass

- `break` stops the loop immediately
- `continue` skips the current iteration
- `pass` does nothing and is used as a placeholder


```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)

for i in range(1, 6):
    if i == 3:
        continue
    print(i)

for i in range(1, 6):
    if i == 3:
        pass
    print(i)
```

## 6. range(), enumerate(), and zip()

These are useful helpers when working with loops.
- `range()` generates a sequence of numbers
- `enumerate()` gives index and value
- `zip()` combines multiple iterables


```python
print(list(range(5)))

fruits = ["apple", "banana", "mango"]
for index, fruit in enumerate(fruits):
    print(index, fruit)

names = ["Ali", "Sara"]
ages = [20, 25]
for name, age in zip(names, ages):
    print(name, age)
```
