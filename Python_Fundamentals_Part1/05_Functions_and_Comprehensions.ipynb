The goal is to make these topics easy to understand with notes and practical examples.


## 1. What is a Function?

A function is a reusable block of code that performs a specific task.

### Why use functions?
- Avoid repetition
- Make code organized
- Make code easier to test and maintain

### Syntax
```python
def function_name(parameters):
    # body
    return value
```


````python
# Simple function
def greet(name):
    return f"Hello, {name}!"

print(greet("Ali"))
print(greet("Sara"))
````

## 2. Function Parameters and Arguments

Functions can accept input values called parameters.

### Types of arguments
- Positional arguments
- Keyword arguments
- Default parameters
- Variable-length arguments


````python
# Positional and keyword arguments
def describe_person(name, age):
    print(f"Name: {name}")
    print(f"Age: {age}")

describe_person("Ali", 21)
describe_person(age=21, name="Ali")

# Default parameters
def greet_user(name="Guest"):
    print(f"Hello, {name}")

greet_user()
greet_user("Sara")

# Variable-length arguments

def add_numbers(*args):
    total = 0
    for value in args:
        total += value
    return total

print(add_numbers(1, 2, 3, 4, 5))
````

## 3. Return Values

A function can return data using the `return` keyword.

### Important points
- A function can return a single value
- It can also return multiple values as a tuple
- If no return is used, it returns `None`


````python
# Returning a single value
def square(x):
    return x * x

print(square(4))

# Returning multiple values
def min_max(values):
    return min(values), max(values)

result = min_max([3, 7, 1, 9])
print(result)
print(type(result))
````

## 4. Scope and Lifetime of Variables

Variables defined inside a function are local to that function.
Variables defined outside are global.

### Key idea
- Local variables exist only inside the function
- Global variables can be accessed from anywhere
- Use `global` carefully when modifying a global variable


````python
# Local vs global scope
x = 10

def show_scope():
    x = 5
    print("Inside function:", x)

show_scope()
print("Outside function:", x)

# Using global
counter = 0

def increase_counter():
    global counter
    counter += 1

increase_counter()
print(counter)
````

## 5. Lambda Functions

A lambda function is a small anonymous function written in one line.

### Syntax
```python
lambda arguments: expression
```

Useful for short operations.


````python
# Lambda function
square = lambda x: x * x
print(square(5))

# Using lambda with sorted()
students = [("Ali", 21), ("Sara", 19), ("Omer", 23)]
students_sorted = sorted(students, key=lambda item: item[1])
print(students_sorted)
````

## 6. Recursion

Recursion is when a function calls itself to solve a smaller version of the same problem.

### Example: factorial


````python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
````

## 7. What are Comprehensions?

Comprehensions provide a short way to create collections.

### Types
- List comprehension
- Dictionary comprehension
- Set comprehension
- Generator expression


````python
# List comprehension
squares = [x * x for x in range(1, 6)]
print(squares)

# List comprehension with condition
evens = [x for x in range(1, 11) if x % 2 == 0]
print(evens)
````

### Dictionary and Set Comprehensions

You can also create dictionaries and sets compactly.


````python
# Dictionary comprehension
square_map = {x: x * x for x in range(1, 6)}
print(square_map)

# Set comprehension
unique_even = {x for x in range(1, 11) if x % 2 == 0}
print(unique_even)
````

## 8. Generator Expressions

Generator expressions are memory-efficient ways to create iterators.

They use `()` instead of `[]`.


````python
# Generator expression
values = (x * x for x in range(1, 6))
print(next(values))
print(next(values))
print(list(values))
````

Try these on your own:
1. Write a function that returns the sum of two numbers.
2. Write a function that takes a list and returns only the even numbers.
3. Use a list comprehension to create squares from 1 to 10.
4. Use a dictionary comprehension to map numbers to their cubes.


````python
# Practice solutions

def add_two(a, b):
    return a + b

print(add_two(3, 4))


def get_even_numbers(values):
    return [x for x in values if x % 2 == 0]

print(get_even_numbers([1, 2, 3, 4, 5, 6]))

squares = [x * x for x in range(1, 11)]
print(squares)

cubes = {x: x * x * x for x in range(1, 6)}
print(cubes)
````
