## 1. Data Types

Python has built-in data types for storing different kinds of values:

- `int` for whole numbers
- `float` for decimal numbers
- `complex` for numbers with real and imaginary parts
- `bool` for True/False values
- `str` for text

### Basic syntax examples

```python
age = 25          # int
price = 19.99     # float
z = 3 + 4j        # complex
is_active = True   # bool
name = "Python"   # str
```

### Checking types

Use `type()` to see the data type of a value:

```python
print(type(age))
print(type(price))
print(type(z))
print(type(is_active))
print(type(name))
```

````python
age = 25
price = 19.99
z = 3 + 4j
is_active = True
name = "Python"

print("age:", age, type(age))
print("price:", price, type(price))
print("z:", z, type(z))
print("is_active:", is_active, type(is_active))
print("name:", name, type(name))
````

## 2. Variables & Memory

Simple words me:

- `Variables` = name jo value ko hold karta hai
- `Multiple Assignment` = ek saath multiple values assign karna
- `Naming Conventions` = variable ka acha aur clear naam rakhna
- `Dynamic Typing` = same variable ki type change ho sakti hai
- `id()` = object ka unique memory address jaisa number
- `Mutable` = object ko change kar sakte hain
- `Immutable` = object ko change nahi kar sakte, naya object banta hai
- `Memory References` = variable actual object ko point karta hai

### Easy examples

```python
x = 10
name = "Ali"

# Multiple assignment
p, q, r = 1, 2, 3

# Good naming
student_name = "Sara"

# Dynamic typing
data = 100
data = "Now I am text"

# Object identity
num1 = 50
num2 = 50
print(id(num1), id(num2))

# Mutable vs immutable
numbers = [1, 2, 3]
numbers.append(4)
text = "hello"
# text[0] = "H"  # not allowed

# Memory references
first_list = [1, 2]
second_list = first_list
second_list.append(3)
```

### Short idea

When you write `second_list = first_list`, both names point to the same list object. If you change one, the other can also show the change.

````python
# Variables
x = 10
name = "Ali"
print("x =", x)
print("name =", name)

# Multiple assignment
p, q, r = 1, 2, 3
print("p, q, r =", p, q, r)

# Naming convention
student_name = "Sara"
print("student_name =", student_name)

# Dynamic typing
data = 100
print("data:", data, type(data))
data = "Now I am text"
print("data:", data, type(data))

# Object identity
num1 = 50
num2 = 50
print("id(num1):", id(num1))
print("id(num2):", id(num2))

# Mutable vs immutable
numbers = [1, 2, 3]
numbers.append(4)
print("numbers:", numbers)

text = "hello"
print("text:", text)

# Memory references
first_list = [1, 2]
second_list = first_list
second_list.append(3)
print("first_list:", first_list)
print("second_list:", second_list)
````
