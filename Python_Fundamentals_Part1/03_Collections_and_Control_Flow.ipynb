# Collections and Flow Control in Python

This notebook covers the core Python collections:
- Lists
- Tuples
- Sets
- Dictionaries

It also explains flow control statements such as:
- if / elif / else
- for loops
- while loops
- break / continue / pass

The focus here is on Python collections as data structures, not external libraries.


## 1. Lists

A list is an ordered, changeable collection of items.

### Key points
- Allows duplicate values
- Uses square brackets `[]`
- Items can be added, removed, or changed

### Example


```python
# Lists
numbers = [10, 20, 30, 40]
print(numbers)
print(numbers[0])
print(numbers[-1])

# Modifying a list
numbers.append(50)
numbers[1] = 25
print(numbers)

# Removing an item
numbers.remove(30)
print(numbers)
```

## 2. Tuples

A tuple is an ordered, unchangeable collection of items.

### Key points
- Uses parentheses `()`
- Immutable, meaning cannot be changed after creation
- Good for fixed data

### Example


```python
# Tuples
point = (10, 20)
print(point)
print(point[0])

# Tuples are immutable
# point[0] = 15  # This would raise an error

# Packing and unpacking
x, y = point
print(x, y)
```

## 3. Sets

A set is an unordered collection of unique items.

### Key points
- Uses curly braces `{}`
- No duplicate values
- Useful for membership testing and removing duplicates

### Example


```python
# Sets
letters = {'a', 'b', 'a', 'c'}
print(letters)

# Add and remove
letters.add('d')
letters.remove('b')
print(letters)

# Check membership
print('a' in letters)
```

## 4. Dictionaries

A dictionary stores data as key-value pairs.

### Key points
- Uses curly braces `{}` with colons `:`
- Keys must be unique
- Great for mapping one value to another

### Example


```python
# Dictionaries
student = {
    'name': 'Ali',
    'age': 21,
    'course': 'Python'
}

print(student)
print(student['name'])
print(student.get('age'))

# Add/update
student['grade'] = 'A'
student['age'] = 22
print(student)

# Remove
student.pop('course')
print(student)
```

## 5. Flow Control

Flow control lets your program make decisions and repeat actions.

### Main concepts
- Conditional statements: `if`, `elif`, `else`
- Loops: `for`, `while`
- Control keywords: `break`, `continue`, `pass`


```python
# Conditional statements
age = 18
if age >= 18:
    print('Adult')
else:
    print('Minor')

# Multiple conditions
score = 75
if score >= 90:
    print('A')
elif score >= 80:
    print('B')
elif score >= 70:
    print('C')
else:
    print('Needs improvement')
```

### For loop

A `for` loop is used to iterate over a sequence.


```python
# For loop over a list
for number in [1, 2, 3, 4]:
    print(number)

# For loop over a string
for char in 'Python':
    print(char)
```

### While loop

A `while` loop repeats as long as the condition is true.


```python
# While loop
count = 1
while count <= 3:
    print(count)
    count += 1
```

### break, continue, and pass

- `break` stops the loop completely
- `continue` skips the current iteration
- `pass` does nothing and acts as a placeholder


```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)

for i in range(1, 6):
    if i == 3:
        continue
    print(i)

for i in range(1, 6):
    if i == 3:
        pass
    print(i)
```
