# Iterators and Built-in Functions in Python

This notebook covers:
- What iterators are
- Iterables vs iterators
- The `iter()` and `next()` functions
- `for` loops with iterators
- Built-in functions such as `len()`, `sum()`, `min()`, `max()`, `sorted()`, and `enumerate()`
- List, tuple, set, and dictionary comprehensions review


## 1. What is an Iterator?

An iterator is an object that can be iterated over one item at a time.

It follows the iterator protocol:
- `iter()` returns an iterator
- `next()` returns the next item


```python
nums = [10, 20, 30]
iterator = iter(nums)

print(next(iterator))
print(next(iterator))
print(next(iterator))
```

## 2. Iterable vs Iterator

An iterable is any object that can return an iterator.
Examples: lists, strings, tuples, dictionaries, sets.

An iterator is the object that actually produces values one by one.


```python
text = "Python"
print(hasattr(text, '__iter__'))
print(hasattr(iter(text), '__next__'))

numbers = [1, 2, 3]
for number in numbers:
    print(number)
```

## 3. Common Built-in Functions

Useful built-in functions include:
- `len()` for length
- `sum()` for total
- `min()` and `max()`
- `sorted()` for sorting
- `enumerate()` for index-value pairs
- `zip()` for combining iterables


```python
values = [5, 2, 9, 1]
print(len(values))
print(sum(values))
print(min(values))
print(max(values))
print(sorted(values))

for index, value in enumerate(values):
    print(index, value)

names = ["Ali", "Sara"]
ages = [20, 25]
for name, age in zip(names, ages):
    print(name, age)
```

## 4. Iterating Over Dictionaries

Dictionaries are iterable, and you can iterate over keys, values, or key-value pairs.


```python
student = {"name": "Ali", "age": 21, "course": "Python"}

print(student.keys())
print(student.values())
print(student.items())

for key in student:
    print(key)

for key, value in student.items():
    print(key, value)
```
