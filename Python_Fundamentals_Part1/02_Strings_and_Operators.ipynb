# Strings and Operators in Python

This notebook covers:
- What strings are
- String creation and indexing
- String methods
- String formatting
- Arithmetic operators
- Comparison operators
- Logical operators
- Assignment operators
- Membership and identity operators


## 1. What is a String?

A string is a sequence of characters enclosed in single quotes, double quotes, or triple quotes.

Examples:
- "Hello"
- 'Python'
- """Multi-line string"""


```python
name = "Python"
message = 'Hello, World!'
multi_line = """This is a
multi-line string"""

print(name)
print(message)
print(multi_line)
```

## 2. String Indexing and Slicing

Each character in a string has an index.
- Indexing starts from 0
- Negative indexing starts from the end
- Slicing extracts a part of the string


```python
text = "Python"
print(text[0])
print(text[-1])
print(text[1:4])
print(text[:3])
print(text[2:])
```

## 3. Common String Methods

Strings come with many built-in methods such as:
- upper(), lower()
- strip(), replace()
- split(), join()
- startswith(), endswith()


```python
text = "  Hello, Python!  "
print(text.upper())
print(text.lower())
print(text.strip())
print(text.replace("Python", "World"))
print(text.split(","))
print("-".join(["a", "b", "c"]))
print(text.startswith("Hello"))
print(text.endswith("!"))
```

## 4. String Formatting

Python offers several ways to format strings:
- f-strings
- format()
- % formatting


```python
name = "Ali"
age = 21
print(f"My name is {name} and I am {age} years old.")
print("My name is {} and I am {} years old.".format(name, age))
print("My name is %s and I am %d years old." % (name, age))
```

## 5. Arithmetic Operators

These operators are used with numbers.
- `+` addition
- `-` subtraction
- `*` multiplication
- `/` division
- `//` floor division
- `%` modulus
- `**` exponentiation


```python
a = 10
b = 3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
```

## 6. Comparison Operators

These compare two values and return True or False.
- `==` equal to
- `!=` not equal to
- `>` greater than
- `<` less than
- `>=` greater than or equal to
- `<=` less than or equal to


```python
x = 5
y = 8
print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= 5)
print(y <= 8)
```

## 7. Logical Operators

These combine boolean conditions.
- `and` → True if both are True
- `or` → True if at least one is True
- `not` → reverses the result


```python
a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

## 8. Assignment Operators

These assign values to variables.
- `=` assigns a value
- `+=`, `-=`, `*=`, `/=` modify a variable


```python
value = 5
value += 3
print(value)
value *= 2
print(value)
```

## 9. Membership and Identity Operators

- `in` checks if an item exists in a sequence
- `not in` checks the opposite
- `is` checks if two variables point to the same object
- `is not` checks the opposite


```python
word = "Python"
print("P" in word)
print("z" not in word)

x = [1, 2, 3]
y = x
z = [1, 2, 3]
print(x is y)
print(x is z)
```
