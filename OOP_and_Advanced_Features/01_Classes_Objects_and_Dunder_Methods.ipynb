# Classes, Objects, and Dunder Methods

This notebook covers the core object-oriented building blocks in Python.

Topics covered:
- defining classes and creating objects
- instance attributes versus class attributes
- methods, `@classmethod`, and `@staticmethod`
- properties for controlled access
- dunder methods like `__repr__`, `__str__`, and `__len__`

## Core Concepts

A class is a blueprint. An object is an instance created from that blueprint.

Python keeps the model lightweight:
- attributes store data
- methods store behavior
- dunder methods customize how objects behave with built-in functions and operators

```python
class BankAccount:
    bank_name = "Python Bank"

    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self._balance = float(balance)

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("deposit amount must be positive")
        self._balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("withdraw amount must be positive")
        if amount > self._balance:
            raise ValueError("insufficient funds")
        self._balance -= amount

    @property
    def balance(self):
        return self._balance

    @classmethod
    def from_string(cls, raw):
        owner, balance = raw.split(":")
        return cls(owner.strip(), float(balance))

    @staticmethod
    def is_valid_amount(amount):
        return isinstance(amount, (int, float)) and amount > 0

    def __repr__(self):
        return f"BankAccount(owner={self.owner!r}, balance={self._balance!r})"

    def __str__(self):
        return f"{self.owner} has ${self._balance:.2f}"

    def __len__(self):
        return int(self._balance)
```

## Example Usage

The example below shows object creation, method calls, and how dunder methods affect printing and built-in functions.

```python
account = BankAccount("Asha", 2500)
account.deposit(500)
account.withdraw(125)

print(account)
print(repr(account))
print(BankAccount.bank_name)
print(account.balance)
print(len(account))

print(BankAccount.from_string("Ravi:1200"))
print(BankAccount.is_valid_amount(42))
```

## Takeaways

Use classes when you want to bundle state and behavior together. Prefer properties when you want controlled access to an attribute. Add dunder methods only when they make the object more natural to use with Python built-ins.
