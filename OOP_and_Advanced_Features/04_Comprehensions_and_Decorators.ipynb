# Comprehensions and Decorators

This notebook covers two Pythonic patterns that appear frequently in real-world libraries.

Topics covered:
- list, dict, and set comprehensions
- nested comprehensions
- generator expressions
- function decorators
- parameterized decorators with `functools.wraps`

## Comprehensions

Comprehensions build collections in a compact and readable form. They usually replace short loops that create a new list, set, or dictionary.

```python
numbers = [1, 2, 3, 4, 5]
squares = [number * number for number in numbers]
even_squares = [number * number for number in numbers if number % 2 == 0]
unique_lengths = {len(word) for word in ["ai", "python", "data"]}
lookup = {number: number * number for number in numbers}

print(squares)
print(even_squares)
print(unique_lengths)
print(lookup)
```

## Decorators

A decorator is a function that takes another function and returns a new function. This is common for logging, validation, memoization, and authorization.

```python
from functools import wraps

def announce(function):
    @wraps(function)
    def wrapper(*args, **kwargs):
        print(f"calling {function.__name__}")
        result = function(*args, **kwargs)
        print("done")
        return result
    return wrapper

@announce
def add(a, b):
    return a + b

print(add(2, 3))
print(add.__name__)
```

## Parameterized Decorator

A decorator can also accept arguments before it wraps the target function.

```python
import time

def repeat(times):
    def decorator(function):
        @wraps(function)
        def wrapper(*args, **kwargs):
            result = None
            for _ in range(times):
                result = function(*args, **kwargs)
            return result
        return wrapper
    return decorator

@repeat(3)
def slow_add(a, b):
    time.sleep(0.1)
    return a + b

print(slow_add(4, 5))
```

## Takeaways

Comprehensions help you write compact collection-building logic. Decorators help you add cross-cutting behavior without changing the original function body.
