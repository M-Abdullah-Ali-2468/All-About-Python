# Iterators, Iterables, and Generators

This notebook explains Python's lazy iteration model and how to build memory-friendly pipelines.

Topics covered:
- iterables versus iterators
- `iter()` and `next()`
- custom iterator classes
- generator functions with `yield`
- generator expressions for compact lazy processing

## Core Concepts

An iterable can produce an iterator. An iterator remembers where it is in a sequence and returns one item at a time. Generators are a short way to create iterators.

```python
class CountDown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        value = self.current
        self.current -= 1
        return value
```

## Iterator Usage

The same object can be advanced with `next()` or used in a `for` loop. Once it is exhausted, it raises `StopIteration`.

```python
counter = CountDown(3)
print(next(counter))
print(next(counter))
print(list(counter))
```

## Generator Functions

A generator function uses `yield` to pause and resume execution. This is useful when data is large or expensive to compute all at once.

```python
def square_numbers(limit):
    for number in range(1, limit + 1):
        yield number * number

print(list(square_numbers(5)))
```

## Generator Expressions

Generator expressions look like list comprehensions, but they produce items lazily. They are ideal when you only need to iterate once.

```python
numbers = (value * value for value in range(1, 4))
for value in numbers:
    print(value)
```

## Takeaways

Use iterators when you want custom step-by-step control. Use generators when you want concise lazy iteration. In data-heavy code, these patterns help keep memory usage low.
