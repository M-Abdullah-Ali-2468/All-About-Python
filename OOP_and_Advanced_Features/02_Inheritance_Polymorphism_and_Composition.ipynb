# Inheritance, Polymorphism, and Composition

This notebook explains how Python classes can reuse, extend, and combine behavior.

Topics covered:
- inheritance and method overriding
- `super()` and cooperative initialization
- polymorphism through shared interfaces
- composition as an alternative to inheritance
- abstract base classes for shared contracts

## Core Concepts

Inheritance lets one class reuse another class. Polymorphism means different objects can respond to the same method call in their own way. Composition means building objects out of other objects instead of forcing a deep class hierarchy.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def speak(self):
        pass

    def describe(self):
        return f"{self.name} is an animal"

class Dog(Animal):
    def speak(self):
        return f"{self.name} says woof"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says meow"
```

## Polymorphism

Any function that depends on the shared interface can work with all subclasses without caring about the concrete type.

```python
def make_it_speak(animal):
    return animal.speak()

animals = [Dog("Buddy"), Cat("Misty")]
for animal in animals:
    print(animal.describe())
    print(make_it_speak(animal))
```

## Composition Example

Composition keeps behavior in separate objects and connects them with attributes. This is often easier to maintain than a large inheritance tree.

```python
class Engine:
    def start(self):
        return "engine started"

class Car:
    def __init__(self, engine):
        self.engine = engine

    def drive(self):
        return f"{self.engine.start()} and car is moving"

car = Car(Engine())
print(car.drive())
```

## Takeaways

Use inheritance when there is a true is-a relationship and a shared contract. Use composition when you want to assemble behavior from smaller reusable parts.
