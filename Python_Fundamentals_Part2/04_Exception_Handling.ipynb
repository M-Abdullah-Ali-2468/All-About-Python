# Exception Handling

Exceptions are runtime errors that interrupt normal program flow. Exception handling lets a program respond to those errors in a controlled way instead of crashing immediately.

## Main keywords
- `try`: code that may fail.
- `except`: code that runs when a matching error happens.
- `else`: code that runs only if no exception occurred.
- `finally`: code that runs no matter what.
- `raise`: manually trigger an exception.

## Why exception handling matters
- It keeps programs from failing unexpectedly.
- It gives users clearer error messages.
- It helps you recover from situations you can predict.
- It supports cleanup logic when resources must be released.

## Common exception types
- `ValueError`
- `TypeError`
- `ZeroDivisionError`
- `FileNotFoundError`
- `IndexError`

## Best practices
- Catch specific exceptions instead of using a broad handler.
- Keep risky code blocks short.
- Use custom exceptions when a built-in one does not describe the problem well.
- Prefer context managers for cleanup when possible.

```python
def safe_inverse(number: float) -> float:
    if number == 0:
        raise ValueError('number must not be zero')
    return 1 / number

for value in [5, 0, -2]:
    try:
        result = safe_inverse(value)
    except ValueError as error:
        print(f'Cannot process {value}: {error}')
    else:
        print(f'Inverse of {value}: {result}')
    finally:
        print(f'Finished processing {value}')
```

```python
class NegativeAgeError(Exception):
    pass

def validate_age(age: int) -> None:
    if age < 0:
        raise NegativeAgeError('age cannot be negative')
    print(f'Valid age: {age}')

try:
    validate_age(21)
    validate_age(-3)
except NegativeAgeError as error:
    print(f'Custom exception caught: {error}')
```

## Summary
Exception handling is a safety layer for Python programs. It improves reliability, keeps error messages understandable, and gives you a structured way to deal with problems as they happen.
