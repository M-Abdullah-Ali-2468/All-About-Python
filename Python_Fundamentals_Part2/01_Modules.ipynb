# Modules

A module is a Python file that groups related code into a reusable unit. Modules help you break a program into smaller parts so the code is easier to read, test, and maintain.

## What a module is
A module can contain functions, classes, constants, and runnable code. When you import a module, Python loads its definitions so you can use them in another file or notebook.

## Why modules matter
- They reduce repetition by letting you reuse code.
- They make programs easier to organize.
- They separate responsibilities into smaller files.
- They support collaboration because different people can work on different modules.

## Types of modules
- Built-in modules such as `math` and `sys`.
- Standard library modules such as `datetime`, `json`, and `pathlib`.
- Third-party modules installed with a package manager.
- User-defined modules that you create yourself.

## Common import patterns
- `import module_name` for importing the full module.
- `import module_name as alias` for shorter names.
- `from module_name import name` for importing specific objects.

## The `__name__` pattern
When a file is imported, `__name__` is the module name. When a file is run directly, `__name__` becomes `__main__`. This lets you keep reusable code and test code in the same file without running everything on import.

```python
import math
import statistics as stats
from datetime import datetime

radius = 4
area = math.pi * radius ** 2
numbers = [12, 18, 24, 30]

print(f'Area of circle: {area:.2f}')
print(f'Mean value: {stats.mean(numbers)}')
print(f'Current timestamp: {datetime.now()}')
print(f'Running under __name__ = {__name__}')
```

## Working with your own module
A custom module is just a normal `.py` file. You import it the same way you import built-in modules. A good module usually focuses on one related topic, such as math helpers, data helpers, or validation utilities.

### Example structure
If you create a file called `helpers.py`, it might contain reusable functions such as `add`, `subtract`, or `format_name`. Another file can then import those functions and use them directly.

### Good module design
- Keep modules focused on one purpose.
- Use clear function and variable names.
- Avoid running heavy code at import time.
- Put direct execution code inside `if __name__ == '__main__':`.

## Summary
Modules are one of the main tools for organizing Python code. They help you build programs that are cleaner, easier to understand, and easier to reuse.
