## 1. Modules

A module is a Python file that contains code you can reuse. Modules help you organize programs into smaller pieces instead of keeping everything in one file.

### Why modules matter
- They improve code reuse.
- They make code easier to test and maintain.
- They reduce repetition by letting you import ready-made functionality.

### Common module types
- Built-in modules: included with Python, such as `math`, `sys`, and `os`.
- Standard library modules: part of the Python distribution, such as `datetime`, `json`, and `pathlib`.
- Third-party modules: installed separately with a package manager.
- User-defined modules: files you create yourself.

### Import styles
- `import module_name`
- `import module_name as alias`
- `from module_name import name1, name2`

### The `__name__` idea
When a file is imported, its `__name__` is the module name. When it is run directly, `__name__` becomes `__main__`. This is useful for separating reusable code from code that should run only when the file is executed directly.

````python
import math
import statistics as stats
from datetime import datetime

radius = 5
area = math.pi * radius ** 2
numbers = [10, 20, 30, 40, 50]

print(f'Area of a circle with radius {radius}: {area:.2f}')
print(f'Mean of {numbers}: {stats.mean(numbers)}')
print(f'Current date and time: {datetime.now()}')
print(f'This notebook runs under __name__ = {__name__}')
````

## 2. Package Managers

A package manager installs, updates, and removes Python packages. The most common package manager in Python is `pip`. It works with packages published on the Python Package Index (PyPI).

### Why package managers matter
- They make it easy to install third-party libraries.
- They help keep project dependencies consistent.
- They work together with virtual environments to avoid version conflicts.

### Important ideas
- Use `python -m pip` instead of calling `pip` directly when possible. This ensures you use the package manager for the Python interpreter you actually want.
- Use virtual environments to isolate project dependencies.
- Use `requirements.txt` to record project dependencies.

### Common commands
```bash
python -m pip install package_name
python -m pip uninstall package_name
python -m pip list
python -m pip freeze > requirements.txt
python -m venv .venv
```

### Conceptual workflow
1. Create a virtual environment.
2. Activate it.
3. Install packages inside that environment.
4. Save dependencies to `requirements.txt`.
5. Recreate the environment later using the same requirements file.

````python
import importlib.util
import sys

print(f'Python executable: {sys.executable}')
print(f'pip module available: {importlib.util.find_spec("pip") is not None}')

try:
    import pip
    print(f'pip version: {pip.__version__}')
except Exception as exc:
    print(f'Could not import pip: {exc}')

print('A requirements file would typically list packages like this:')
print('requests==2.32.3')
print('pandas==2.2.2')
````

## 3. File I/O

File I/O means reading from and writing to files. Python can work with text files, binary files, CSV files, JSON files, and many other formats.

### Core concepts
- **Read mode (`r`)** opens an existing file for reading.
- **Write mode (`w`)** creates a new file or overwrites an existing one.
- **Append mode (`a`)** adds content to the end of a file.
- **Binary mode (`b`)** is used for non-text data such as images or audio.
- **Context managers** (`with`) automatically close files when you are done.

### Good practices
- Always prefer `with open(...)` so files close safely.
- Choose the correct encoding, usually `utf-8`.
- Use `pathlib.Path` for cleaner file path handling.
- Keep file operations small and focused.

### Typical text workflow
1. Open the file.
2. Read or write content.
3. Close the file automatically with `with`.

````python
from pathlib import Path
from tempfile import TemporaryDirectory

sample_text = "Modules, package managers, file I/O, and exceptions are core Python skills."

with TemporaryDirectory() as temp_dir:
    temp_path = Path(temp_dir) / 'sample_notes.txt'

    with temp_path.open('w', encoding='utf-8') as file_handle:
        file_handle.write(sample_text)

    with temp_path.open('r', encoding='utf-8') as file_handle:
        loaded_text = file_handle.read()

    print(f'Created file: {temp_path.name}')
    print(f'Read back text: {loaded_text}')
    print(f'Line count: {len(loaded_text.splitlines())}')

print('Temporary files are removed automatically when the context ends.')
````

## 4. Exception Handling

Exceptions are runtime errors that interrupt normal program flow. Exception handling lets you respond to those errors gracefully instead of letting the program crash immediately.

### Why exception handling matters
- It prevents unexpected crashes.
- It lets you display useful error messages.
- It allows recovery when a problem can be handled safely.

### Main keywords
- `try`: code that might fail.
- `except`: code that runs when an error happens.
- `else`: code that runs only if no exception occurs.
- `finally`: code that runs no matter what.
- `raise`: manually trigger an exception.

### Common exception types
- `ValueError`
- `TypeError`
- `ZeroDivisionError`
- `FileNotFoundError`
- `IndexError`

### Best practices
- Catch only the exceptions you expect.
- Avoid using a bare `except` unless you have a strong reason.
- Keep the risky code block as small as possible.
- Use `finally` or context managers when cleanup is needed.

````python
def safe_inverse(number: float) -> float:
    if number == 0:
        raise ValueError('number must not be zero')
    return 1 / number

for value in [4, 0, -2]:
    try:
        result = safe_inverse(value)
    except ValueError as error:
        print(f'Cannot process {value}: {error}')
    else:
        print(f'Inverse of {value} is {result}')
    finally:
        print(f'Finished processing {value}')

class NegativeAgeError(Exception):
    pass

def validate_age(age: int) -> None:
    if age < 0:
        raise NegativeAgeError('age cannot be negative')
    print(f'Valid age: {age}')

try:
    validate_age(25)
    validate_age(-3)
except NegativeAgeError as error:
    print(f'Custom exception caught: {error}')
````
