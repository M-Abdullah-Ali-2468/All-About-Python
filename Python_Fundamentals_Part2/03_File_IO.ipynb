# File I/O

File I/O means reading data from files and writing data back to files. It is how Python programs store information outside memory so that the data can be used again later.

## Core concepts
- Read mode (`r`) opens an existing file for reading.
- Write mode (`w`) creates or overwrites a file.
- Append mode (`a`) adds new content to the end of a file.
- Binary mode (`b`) is used for non-text data.
- The `with` statement closes files automatically.

## Why this matters
- Programs can save results for later use.
- Text files can hold notes, logs, and configuration.
- JSON and CSV files are common for structured data.
- Clean file handling prevents resource leaks and broken file operations.

## Good practices
- Prefer `pathlib.Path` for paths.
- Use `utf-8` encoding for text files.
- Keep file operations inside a context manager.
- Handle missing files and bad input gracefully.

```python
from pathlib import Path
from tempfile import TemporaryDirectory

text = 'First line\nSecond line\nThird line'

with TemporaryDirectory() as temp_dir:
    file_path = Path(temp_dir) / 'notes.txt'

    with file_path.open('w', encoding='utf-8') as handle:
        handle.write(text)

    with file_path.open('r', encoding='utf-8') as handle:
        content = handle.read()

    print(f'File name: {file_path.name}')
    print('Content read from file:')
    print(content)
    print(f'Line count: {len(content.splitlines())}')
```

## Common file formats
Text files are the simplest form of file storage, but many real projects use structured formats such as CSV and JSON. Python has built-in support for JSON and good support for CSV, which makes those formats easy to work with.

### Typical use cases
- Log files for debugging and auditing.
- Configuration files for application settings.
- Data files for exchanging structured records.
- Reports and exported results.

## Summary
File I/O lets Python programs persist and retrieve data. Using context managers, explicit encodings, and good path handling makes file code safer and easier to maintain.
