# Package Managers

A package manager installs, updates, removes, and tracks Python packages. In everyday Python work, the most common package manager is `pip`. Package managers are essential when you need third-party libraries that are not included in the standard library.

## Why package managers matter
- They make dependency installation easy.
- They let projects depend on the same package versions.
- They support reproducible environments.
- They work best when combined with virtual environments.

## Core ideas
- Install a package with `python -m pip install package_name`.
- Remove a package with `python -m pip uninstall package_name`.
- List installed packages with `python -m pip list`.
- Save dependencies with `python -m pip freeze > requirements.txt`.

## Virtual environments
A virtual environment creates an isolated Python setup for a project. This prevents one project from breaking another by using incompatible package versions.

### Typical workflow
1. Create the environment.
2. Activate the environment.
3. Install the packages you need.
4. Save the installed versions.
5. Recreate the same environment later using the saved list.

```python
import importlib.util
import sys

print(f'Python executable: {sys.executable}')
print(f'pip available: {importlib.util.find_spec("pip") is not None}')

try:
    import pip
    print(f'pip version: {pip.__version__}')
except Exception as error:
    print(f'Could not import pip: {error}')

print('Example requirements entry:')
print('requests==2.32.3')
```

## Project dependency files
A `requirements.txt` file records the exact packages and versions used in a project. This is useful when another person needs to install the same dependencies or when you want to recreate the same environment later.

### Notes on dependency management
- Lock files and pinned versions improve reproducibility.
- Isolate dependencies per project whenever possible.
- Keep dependency lists small and only install what the project really uses.
- Review package names carefully before installing anything from PyPI.

## Summary
Package managers are the standard way to bring external code into your Python projects. They are most effective when paired with virtual environments and a saved dependency list.
